```python
import pandas as pd
import glob
```

```python
news_files = sorted(glob.glob("./CMIN-Dataset/CMIN-US/news/raw/*.csv"))
```

```python
aapl_news = pd.read_csv(news_files[0], delimiter="\t")
aapl_news = aapl_news.iloc[::-1].reset_index().drop('index', axis=1)
aapl_news['date'] = pd.to_datetime(aapl_news['date'])
aapl_news['time'] = pd.to_datetime(aapl_news['time'])
aapl_news
```

```python
dates = sorted(tuple(set(aapl_news['date'])))
len(dates)
```

```python
date = '2019-01-22'
len(aapl_news[aapl_news['date'] == date])
```

```python
aapl_now = aapl_news[aapl_news['date'] == date]

for i in range(len(aapl_now)):
    title = aapl_now['title'].iloc[i]
    summary = aapl_now['summary'].iloc[i]
    
    print("##", title)
    print()
    print(summary)
    print()
    print()
```
